```python
import numpy as np 
import pandas as pd 
```

```python
df=pd.read_csv('zepto_mess.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Items count should be positive

df = df[df['Items_Count'] > 0]
```

```python
# Distance should be realistic (0–20 km)

df = df[(df['Distance_km'] > 0) & (df['Distance_km'] < 20)]
```

```python
# Delivery time realistic (0–60 mins)

df = df[(df['Delivery_Time'] > 0) & (df['Delivery_Time'] < 60)]
```

```python
# Delivery charge >= 0

df = df[df['Delivery_Charge'] >= 0]
```

```python
df.isnull().sum()
```

```python
df = df[(df['Customer_Rating'].between(1,5)) & 
        (df['Rider_Rating'].between(1,5))]
```

```python
df['Surge_Pricing'] = df['Surge_Pricing'].apply(lambda x: 1 if x==1 else 0)
```

```python
df.replace(['??','Unknown','Crypto?'], np.nan, inplace=True)
```

```python
# num_cols = [
#     'Delivery_Time',
#     'Order_Value',
#     'Items_Count',
#     'Distance_km',
#     'Delivery_Charge',
#     'Rider_Rating',
#     'Customer_Rating'
# ]

# for i in num_cols:
#     df[i].fillna(df[i].median(), inplace=True)
```

```python
cat_cols = [
    'Time_Slot',
    'Warehouse_Location',
    'Payment_Method',
    'Inventory_Issue',
    'Order_Status'
]

for i in cat_cols:
    df[i].fillna(df[i].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().any()
```

```python
df['Delivery_Efficiency'] = df['Distance_km'] / df['Delivery_Time']
```

```python
df.head()
```

```python
df['Avg_Order_Value'] = df['Order_Value'] / df['Items_Count']
```

```python
df['Late_Delivery'] = (df['Delivery_Time'] > 20).astype(int)
```

```python
# df['CX_Score'] = (df['Customer_Rating'] + df['Rider_Rating']) / 2
```

```python
df['Cancelled'] = (df['Order_Status'] == 'Cancelled').astype(int)
```

```python
df['Net_Revenue'] = df['Order_Value'] - df['Delivery_Charge']
```

```python
df['Order_Size'] = pd.cut(df['Order_Value'], 3, labels=['Low','Medium','High'])
```

```python
df['Distance_Category'] = pd.cut(df['Distance_km'],
                                bins=[0,5,10,20],
                                labels=['Near','Medium','Far'])
```

```python
df.head()
```

```python
df.to_csv("zepto_cleaned.csv", index=False)
```
